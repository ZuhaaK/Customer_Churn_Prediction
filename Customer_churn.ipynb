```python
#Step 7.3 Grid search to tune Logistic Regression
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression

# We'll use a LogisticRegression with class_weight='balanced' to keep focus on churners.
base_clf = LogisticRegression(class_weight='balanced', max_iter=5000, random_state=42)

# Parameter grid to search
param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],   # regularization strength (smaller C -> stronger regularization)
    'penalty': ['l1', 'l2'],        # l1 for sparsity (zeroing some coeffs), l2 for ridge-like shrinkage
    'solver': ['liblinear']         # liblinear supports l1 and l2 for binary problems and is stable here
}

# GridSearchCV setup:
# - scoring='f1' optimizes for F1 score (balance of precision & recall)
# - cv=5 does 5-fold cross-validation on training data
# - n_jobs=-1 uses all CPU cores for speed
grid = GridSearchCV(estimator=base_clf,
                    param_grid=param_grid,
                    scoring='f1',
                    cv=5,
                    n_jobs=-1,
                    verbose=1)

# Fit grid search on the scaled training data
grid.fit(X_train_scaled, y_train)

# Best parameters and best CV score
print("Best params:", grid.best_params_)
print("Best CV F1:", grid.best_score_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import numpy as np

# 1ï¸âƒ£ Polynomial Features (interaction only)
poly = PolynomialFeatures(degree=2, interaction_only=True, include_bias=False)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)

# 2ï¸âƒ£ Recursive Feature Elimination (select top 10 features)
rfe = RFE(LogisticRegression(max_iter=5000, solver='liblinear'), n_features_to_select=10)
rfe.fit(X_train_poly, y_train)

selected_features_idx = rfe.support_
print("Selected Polynomial Feature Indices:", np.where(selected_features_idx)[0])

X_train_rfe = X_train_poly[:, selected_features_idx]
X_test_rfe = X_test_poly[:, selected_features_idx]

# 3ï¸âƒ£ Train Logistic Regression on selected features
final_log_reg = LogisticRegression(max_iter=5000, class_weight='balanced', solver='liblinear', C=1)
final_log_reg.fit(X_train_rfe, y_train)

# 4ï¸âƒ£ Predict probabilities
y_pred_proba = final_log_reg.predict_proba(X_test_rfe)[:, 1]

# 5ï¸âƒ£ Threshold tuning: Lower threshold to 0.4 to catch more churners
y_pred_threshold = np.where(y_pred_proba >= 0.4, 1, 0)

# 6ï¸âƒ£ Evaluate
print("\nAccuracy:", accuracy_score(y_test, y_pred_threshold))
print("\nClassification Report:\n", classification_report(y_test, y_pred_threshold))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_threshold))
```

```python
# Model setup catches nearly all churners (high recall) by lowering threshold.
# Some non-churners get flagged (lower precision), so overall accuracy drops.
# Using only 10 key features makes the model simpler and easier to explain.
#In churn prediction, catching potential churners is more important than overall accuracy
```

```python
#Hyperparameter Tuning

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.05, 0.1],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0],
    'scale_pos_weight': [1, sum(y_train==0)/sum(y_train==1)]  # handle imbalance
}

grid_xgb = GridSearchCV(
    estimator=xgb_clf,
    param_grid=param_grid,
    scoring='recall',   # focus on catching churners
    cv=3,
    n_jobs=-1,
    verbose=2
)

# Fit to training data
grid_xgb.fit(X_train_scaled, y_train)

# Best parameters
print("Best Parameters:", grid_xgb.best_params_)
print("Best Recall Score:", grid_xgb.best_score_)


```

```python
#Explanation

# Define the hyperparameter grid:
# - n_estimators: number of trees in the ensemble (more trees can improve learning but may overfit)
# - max_depth: maximum depth of each tree (controls complexity)
# - learning_rate: step size shrinkage to prevent overfitting (smaller values = slower learning, better generalization)
# - subsample: fraction of training samples used per tree (helps prevent overfitting)
# - colsample_bytree: fraction of features used per tree (reduces correlation between trees)
# - scale_pos_weight: balances classes if the dataset is imbalanced (gives more weight to minority class)
#   - Here, we try 1 (no weighting) and the ratio of majority/minority class in training data

# Create a GridSearchCV object to find the best hyperparameters:
# - estimator: the XGBoost model we defined earlier
# - param_grid: the hyperparameter combinations to test
# - scoring='recall': we focus on recall to catch as many churners as possible
# - cv=3: 3-fold cross-validation splits the training data into 3 parts for more reliable evaluation
# - n_jobs=-1: use all CPU cores for faster computation
# - verbose=2: prints progress of which parameter combination is being tested

# Fit the GridSearchCV object on the training data:
# - Trains the model for all combinations of hyperparameters using 3-fold CV
# - Calculates recall for each combination
# - Stores results so we can select the best performing combination

# After fitting:
# - grid_xgb.best_params_ gives the combination of hyperparameters with the highest recall
# - grid_xgb.best_score_ gives the mean recall across the 3 CV folds for the best combination

GridSearchCV tested all hyperparameter combinations using 3-fold CV.

Best parameters were found with very high recall (~0.83).

Model is now ready for threshold tuning, predictions, and evaluation.
```

```python
import plotly.graph_objects as go

# Metrics for each model
metrics_dict = {
    'Logistic Regression': [0.72779, 0.492537, 0.794118, 0.607984, 0.830522],
    'Random Forest': [0.778962, 0.567452, 0.708556, 0.630202, 0.839460],
    'XGBoost': [0.745558, 0.514085, 0.780749, 0.619958, 0.835809],
    'LightGBM': [0.744847, 0.513612, 0.756684, 0.611892, 0.832125]
}

metrics_labels = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']

# Create radar chart
fig = go.Figure()

for model, values in metrics_dict.items():
    fig.add_trace(go.Scatterpolar(
        r=values + [values[0]],   # Close the loop
        theta=metrics_labels + [metrics_labels[0]],
        fill='toself',
        name=model
    ))

fig.update_layout(
    polar=dict(
        radialaxis=dict(visible=True, range=[0, 1])
    ),
    title='Interactive Radar Chart - Model Comparison',
    showlegend=True
)

fig.show()
```

```python
import plotly.graph_objects as go

# Models and their metrics
models = ['Logistic Regression', 'Random Forest', 'XGBoost', 'LightGBM']

accuracy = [0.72779, 0.778962, 0.745558, 0.744847]
precision = [0.492537, 0.567452, 0.514085, 0.513612]
recall = [0.794118, 0.708556, 0.780749, 0.756684]
f1 = [0.607984, 0.630202, 0.619958, 0.611892]
roc_auc = [0.830522, 0.839460, 0.835809, 0.832125]

# Create figure
fig = go.Figure()

# Add traces for each metric
fig.add_trace(go.Bar(x=models, y=accuracy, name='Accuracy'))
fig.add_trace(go.Bar(x=models, y=precision, name='Precision'))
fig.add_trace(go.Bar(x=models, y=recall, name='Recall'))
fig.add_trace(go.Bar(x=models, y=f1, name='F1-Score'))
fig.add_trace(go.Bar(x=models, y=roc_auc, name='ROC-AUC'))

# Update layout
fig.update_layout(
    barmode='group',       # Group bars side by side
    title='Interactive Comparison of All Models',
    xaxis_title='Models',
    yaxis_title='Metric Values',
    yaxis=dict(range=[0, 1]),  # All metrics scaled 0-1
    legend_title='Metrics'
)

fig.show()
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Models and metrics
models = ['Logistic Regression', 'Random Forest', 'XGBoost', 'LightGBM']
metrics_names = ['Accuracy', 'Precision', 'Recall', 'F1-Score', 'ROC-AUC']
metrics_values = {
    'Logistic Regression': [0.72779, 0.492537, 0.794118, 0.607984, 0.830522],
    'Random Forest': [0.778962, 0.567452, 0.708556, 0.630202, 0.839460],
    'XGBoost': [0.745558, 0.514085, 0.780749, 0.619958, 0.835809],
    'LightGBM': [0.744847, 0.513612, 0.756684, 0.611892, 0.832125]
}

# Create subplots: 1 row, 2 columns (radar + bar)
fig = make_subplots(
    rows=1, cols=2,
    specs=[[{'type': 'polar'}, {'type': 'xy'}]],
    subplot_titles=("Radar Plot of Model Metrics", "Bar Plot of F1-Scores")
)

# Add radar traces
for model in models:
    fig.add_trace(go.Scatterpolar(
        r=metrics_values[model],
        theta=metrics_names,
        fill='toself',
        name=model
    ), row=1, col=1)

# Add bar plot traces for F1-Score comparison
f1_scores = [metrics_values[m][3] for m in models]  # index 3 is F1-Score
fig.add_trace(go.Bar(
    x=models,
    y=f1_scores,
    text=[f"{s:.3f}" for s in f1_scores],
    textposition='auto',
    marker_color='teal',
    name='F1-Score'
), row=1, col=2)

# Update layout
fig.update_layout(
    height=500, width=1000,
    title_text="Model Comparison Dashboard: Radar + F1-Score Bar",
    showlegend=True,
    polar=dict(radialaxis=dict(visible=True, range=[0,1]))
)

fig.show()
```
